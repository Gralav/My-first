```python
# =================================================================
# WABANAKI HEALTH NEEDS ASSESSMENT (WHNA) 2022: MASTER ANALYSIS SCRIPT
# =================================================================
# This script is a full-pipeline implementation of your research plan.
# STRATEGY: Analyze the "Health of the Data" (Time/Completion) on
# RAW data first, then perform cleaning and deep clinical indexing.
# =================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from scipy.stats import pearsonr
import os
```

```python
# -----------------------------------------------------------------
# BLOCK 1: DATA LOADING AND SYSTEM AUDIT
# EXPLANATION:
# This block loads the raw CSV and identifies which columns are
# "Completion Markers" and which are "Data Variables".
# We also define the mapping dictionaries for Tribal Nations and
# Gender to ensure all charts are human-readable from the start.
# -----------------------------------------------------------------
filename = 'whna2022 data_withcomputedvar_20240503 (1).csv'
# Added encoding='latin1' to handle UnicodeDecodeError
df_raw = pd.read_csv(filename, encoding='latin1')

# Mapping Dictionaries (Based on Codebook)
tribe_map = {1.0: 'Penobscot', 2.0: 'Passamaquoddy (PP)', 3.0: 'Passamaquoddy (IT)', 4.0: 'Houlton Maliseet', 5.0: 'Micmac'}
gender_map = {1.0: 'Male', 2.0: 'Female', 3.0: 'Non-binary', 4.0: 'Transgender', 5.0: 'Self-describe'}

# Identify Completion indicators vs. Data variables
complete_cols = [col for col in df_raw.columns if col.endswith('_complete')]
data_cols = [col for col in df_raw.columns if not col.endswith('_complete')]

print(f"Audit: Found {len(complete_cols)} Thematic Sections and {len(df_raw)} raw submissions.")
```

```python
# Ensure 'whna_2022_date' is datetime object
df_raw['whna_2022_date'] = pd.to_datetime(df_raw['whna_2022_date'], errors='coerce')

# Drop rows where date is NaN after conversion
df_time_analysis = df_raw.dropna(subset=['whna_2022_date']).copy()

# Extract time components
df_time_analysis['day_of_week'] = df_time_analysis['whna_2022_date'].dt.day_name()
df_time_analysis['hour_of_day'] = df_time_analysis['whna_2022_date'].dt.hour
df_time_analysis['week_number'] = df_time_analysis['whna_2022_date'].dt.isocalendar().week

# Map gender for readability
df_time_analysis['Gender_Label'] = df_time_analysis['gender'].map(gender_map)

print("Extracted time components and mapped gender for analysis.")
```

```python
# -----------------------------------------------------------------
# BLOCK 3: COMPLETION & PARTICIPATION BEHAVIOR (RESEARCH PLAN SEC 2)
# EXPLANATION:
# We analyze completion rates across the RAW data to identify
# drop-off points. This visualizes which sections (themes) were
# successfully finished by different Tribal Communities and where
# participants may have lost interest or found the survey too long.
# -----------------------------------------------------------------
print("\n[Section 2] Analyzing Thematic Completion Rates (Raw Data)...")
theme_cols = [c for c in complete_cols if c != 'whna_2022_complete']
comp_melt = df_raw.melt(id_vars=['tribnation'], value_vars=theme_cols,
                        var_name='Theme', value_name='Status')
comp_melt['Is_Complete'] = comp_melt['Status'] == 2
comp_melt['Community'] = comp_melt['tribnation'].map(tribe_map)

# Calculate completion rate per theme for each community
theme_community_completion_rates = comp_melt.groupby(['Theme', 'Community'])['Is_Complete'].mean().reset_index()

plt.figure(figsize=(15, 8)) # Increased figure size for better readability with more bars
sns.barplot(data=theme_community_completion_rates, x='Theme', y='Is_Complete', hue='Community', palette='tab10')
plt.xticks(rotation=45, ha='right')
plt.title('Section Completion Rate by Theme and Community')
plt.ylabel('Completion Proportion')
plt.ylim(0, 1) # Ensure y-axis is from 0 to 1 for completion proportion
plt.legend(title='Tribal Nation', bbox_to_anchor=(1.05, 1), loc='upper left') # Move legend outside to prevent overlap
plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

```python
# -----------------------------------------------------------------
# BLOCK 4: DATA CLEANING & QUALITY CONTROL (RESEARCH PLAN SEC 1)
# EXPLANATION:
# Now that we understand the raw patterns, we refine the data:
# 1. Drop columns that are 80% empty to remove "noise".
# 2. Filter for "Fully Complete" records (where all markers == 2).
# 3. Deduplicate by keeping the entry with the most information.
# 4. Replace '555' (Prefer not to answer) with NaN for clean averages.
# -----------------------------------------------------------------
print("\n[Section 3] Executing Data Cleaning and Filtering...")

# 1. Remove sparse columns
threshold = 0.80
limit = len(df_raw) * (1 - threshold)
df_reduced = df_raw.dropna(axis=1, thresh=limit).copy()

# 2. Filter for Fully Complete records
is_fully_complete = (df_reduced[complete_cols] == 2).all(axis=1)
df_complete = df_reduced[is_fully_complete].copy()
df_incomplete = df_reduced[~is_fully_complete].copy()

# 3. Intelligent Deduplication
df_complete['null_count'] = df_complete.isnull().sum(axis=1)
df_complete = df_complete.sort_values(by=['whna_2022_id', 'null_count'])
df_clean = df_complete.drop_duplicates(subset='whna_2022_id', keep='first').copy()
df_clean = df_clean.drop(columns=['null_count'])

# 4. Handle Special Codes
df_clean = df_clean.replace(555, np.nan)

# Save intermediate files
df_incomplete.to_csv('incomplete_records.csv', index=False)
df_clean.to_csv('fully_cleaned_complete_data.csv', index=False)

print(f"Cleaning complete. {len(df_clean)} high-quality records kept for deep analysis.")
```

```python
# -----------------------------------------------------------------
# BLOCK 5: CONSTRUCTING THE CULTURAL CONNECTEDNESS INDEX (SEC 3)
# EXPLANATION:
# We build the 'Outcome' variable by summing variables related to
# Tribal connection and language. This index represents a
# strength-based protective factor. We then divide participants
# into quartiles based on their connection strength.
# -----------------------------------------------------------------
print("\n[Section 4] Building the Cultural Connectedness Index...")
conn_vars = ['comm_tribe', 'you_tribe', 'mem_tribe', 'language']
df_clean['cultural_conn_score'] = df_clean[conn_vars].sum(axis=1)

# Categorize into quartiles
df_clean['conn_quartile'] = pd.qcut(df_clean['cultural_conn_score'], 4,
                                   labels=['Q1-Low Connection', 'Q2', 'Q3', 'Q4-High Connection'])

plt.figure(figsize=(10, 5))
sns.histplot(df_clean['cultural_conn_score'], kde=True, color='darkorchid')
plt.title('Distribution of Cultural Connectedness Scores')
plt.show()
```

```python
# -----------------------------------------------------------------
# BLOCK 6: CORRELATION AND INFLUENCES (SEC 4)
# EXPLANATION:
# We test the relationship between Cultural Connection, Stressors
# (Discrimination/Loss), and Mental Health (PHQ-8/GAD-7).
# This heatmap reveals if identity acts as a buffer against
# the negative health impacts of social stressors.
# -----------------------------------------------------------------
print("\n[Section 5] Testing Stress and Mental Health Interactions...")

# PHQ-8 and GAD-7 totals
phq_items = [c for c in df_clean.columns if 'phq_' in c]
gad_items = [c for c in df_clean.columns if 'gad_' in c]
df_clean['phq_total'] = df_clean[phq_items].sum(axis=1)
df_clean['gad_total'] = df_clean[gad_items].sum(axis=1)

# Discrimination and Historical Loss Totals
disc_cols = [c for c in df_clean.columns if c.startswith('iden_')]
loss_cols = [c for c in df_clean.columns if c.startswith('loss_')]
df_clean['discrim_score'] = df_clean[disc_cols].sum(axis=1)
df_clean['loss_score'] = df_clean[loss_cols].sum(axis=1)

# Generate Correlation Heatmap
corr_list = ['cultural_conn_score', 'phq_total', 'gad_total', 'discrim_score', 'loss_score', 'age']
plt.figure(figsize=(10, 8))
sns.heatmap(df_clean[corr_list].corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix: Connectedness vs. Health Stressors')
plt.show()
```

```python
# -----------------------------------------------------------------
# BLOCK 7: THEMATIC VISUALIZATIONS (BMI & HEALTH)
# EXPLANATION:
# Using the cleaned data, we visualize BMI by Gender and General
# Health by Community. This highlights disparities and identifies
# where Tribal health programs should target physical health outreach.
# -----------------------------------------------------------------
print("\n[Section 6] Analyzing BMI and General Health Status...")

# 1. BMI by Gender
if 'gen_weigh' in df_clean.columns:
    df_clean['height_in'] = (df_clean['gen_tall_ft'] * 12) + df_clean['gen_tall_in']
    df_clean['bmi'] = (df_clean['gen_weigh'] * 703) / (df_clean['height_in']**2)
    # Clip realistic range to remove data errors
    bmi_plot_df = df_clean[(df_clean['bmi'] >= 15) & (df_clean['bmi'] <= 60)].copy()
    bmi_plot_df['Gender'] = bmi_plot_df['gender'].map(gender_map)

    plt.figure(figsize=(10, 6))
    sns.violinplot(x='Gender', y='bmi', data=bmi_plot_df, inner='quartile', palette='Set3')
    plt.title('BMI Distribution by Gender Identity')
    plt.show()

# 2. General Health Status by Community
health_labels = {1.0: 'Excellent', 2.0: 'Very Good', 3.0: 'Good', 4.0: 'Fair', 5.0: 'Poor'}
df_clean['Health_Label'] = df_clean['gen_heal'].map(health_labels)
plt.figure(figsize=(12, 6))
sns.countplot(data=df_clean, x='Health_Label', hue='tribnation',
              order=['Excellent', 'Very Good', 'Good', 'Fair', 'Poor'])
plt.title('Self-Rated General Health by Community')
plt.legend(title='Tribe', labels=tribe_map.values(), bbox_to_anchor=(1.05, 1))
plt.show()
```

```python
# -----------------------------------------------------------------
# BLOCK 8: WORDCLOUDS FOR QUALITATIVE INSIGHTS
# EXPLANATION:
# We scan for every column containing text responses (narratives)
# and generate a WordCloud for each. This provides a visual summary
# of community strengths, health concerns, and risky behaviors
# expressed in the participants' own words.
# -----------------------------------------------------------------
print("\n[Section 7] Generating Qualitative Thematic WordClouds...")
text_cols = df_clean.select_dtypes(include=['object']).columns
# Exclude metadata columns
filtered_text_cols = [c for c in text_cols if not any(k in c.lower() for k in ['id', 'date', 'complete', 'desc'])]

for col in filtered_text_cols:
    combined_text = " ".join(df_clean[col].dropna().astype(str).tolist())
    if len(combined_text.strip()) > 30:
        print(f"Creating WordCloud for: {col}")
        wc = WordCloud(width=800, height=400, background_color='white',
                       colormap='plasma', max_words=50).generate(combined_text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wc, interpolation='bilinear')
        plt.axis('off')
        plt.title(f"Thematic WordCloud: {col.replace('_', ' ').title()}", fontsize=14)
        plt.show()

# --- FINAL OUTPUT ---
df_clean.to_csv('WHNA_2022_Final_Cleaned_Analysis.csv', index=False)
print("\nSUCCESS: Analysis complete. Final data saved to 'WHNA_2022_Final_Cleaned_Analysis.csv'.")
```
