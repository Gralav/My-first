```python
import pandas as pd

# Load the dataset
file_path = 'Mine.csv'
df = pd.read_csv(file_path)

# Inspect the data
print("First 5 rows of the data:")
print(df.head())
print("\nData Info:")
print(df.info())
print("\nColumn names:")
print(df.columns.tolist())
```

```python
# Summary statistics for numerical columns
desc = df.describe()

# Check for missing values (count and percentage)
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({'Missing Count': missing_values, 'Percentage': missing_percentage})
missing_df_sorted = missing_df.sort_values(by='Percentage', ascending=False)

print("Summary Statistics (first 10 numerical columns):")
print(desc.iloc[:, :10])

print("\nTop 20 columns with most missing values:")
print(missing_df_sorted.head(20))

# Check unique values for some demographics
print("\nUnique values for 'bio_sex':", df['bio_sex'].unique())
print("Unique values for 'gender':", df['gender'].unique())
print("Unique values for 'age':", df['age'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Clean age for plotting (exclude 555)
clean_age = df[df['age'] <= 120]['age']

# Plotting Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(clean_age, bins=20, kde=True, color='skyblue')
plt.title('Age Distribution (excluding codes)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('age_distribution.png')

# Plotting bio_sex and gender distributions (excluding 555)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(x='bio_sex', data=df[df['bio_sex'] < 10], ax=axes[0], palette='viridis')
axes[0].set_title('Biological Sex Counts')

sns.countplot(x='gender', data=df[df['gender'] < 10], ax=axes[1], palette='magma')
axes[1].set_title('Gender Counts')

plt.tight_layout()
plt.savefig('demographics_counts.png')

# Summary of Chronic Conditions
chronic_cols = ['chron_obesity', 'chron_dm', 'chron_pdm', 'chron_heart', 'chron_hbp', 'chron_cancer']
chronic_summary = df[chronic_cols].apply(pd.Series.value_counts).transpose()
print("\nChronic Conditions Summary (Counts):")
print(chronic_summary)
```

```python
# tribnation distribution
trib_counts = df['tribnation'].value_counts()
print("Tribnation Counts:\n", trib_counts)

# Identify Perceived Discrimination and Historical Loss columns
discrim_cols = [col for col in df.columns if col.startswith('iden_')]
loss_cols = [col for col in df.columns if col.startswith('loss_')]

# Check value counts for a few
print("\nExample Perceived Discrimination (iden_unfair):")
print(df['iden_unfair'].value_counts())

print("\nExample Historical Loss (loss_land):")
print(df['loss_land'].value_counts())

# Look at topthree_healthissues text
print("\nSample entries for 'topthree_healthissues':")
print(df['topthree_healthissues'].dropna().head(10))
```

```python
# Calculate mean for discrimination and loss columns (excluding 555)
discrim_means = df[discrim_cols].replace(555, pd.NA).mean()
loss_means = df[loss_cols].replace(555, pd.NA).mean()

print("Average Perceived Discrimination scores (lower/higher scale needs context):")
print(discrim_means.head())

print("\nAverage Historical Loss scores:")
print(loss_means.head())

# Health vs Tribnation
health_by_tribe = df.groupby('tribnation')['comm_health'].mean()
print("\nMean Community Health Rating by Tribnation (lower might be better or worse depending on scale):")
print(health_by_tribe)
```

```python
# Clean comm_health and re-calculate by tribnation
clean_comm_health = df['comm_health'].replace(555, pd.NA)
health_by_tribe_clean = df.assign(comm_health_clean=clean_comm_health).groupby('tribnation')['comm_health_clean'].mean()

print("Cleaned Mean Community Health Rating by Tribnation:")
print(health_by_tribe_clean)

# Value counts for comm_health to see the scale
print("\nValue counts for comm_health (cleaned):")
print(clean_comm_health.value_counts().sort_index())
```

```python
# --- SECTION 2: SURVEY QUALITY & COMPLETION ANALYSIS ---

# 2.1 Response & Completion Patterns
completion_cols = [col for col in df.columns if col.endswith('_complete')]
# Assuming 2 = Complete, 1 = Unverified/Partial, 0 = Incomplete
comp_rates = (df[completion_cols] == 2).mean() * 100

# Drop-off analysis (cumulative completion)
# We assume the columns appear in chronological order in the survey
ordered_sections = [
    'basic_demographics_complete',
    'community_health_and_strengths_complete',
    'connection_to_community_and_culture_complete',
    'perceived_discrimination_and_historical_loss_complete',
    'health_and_healthcare_access_complete',
    'physical_activity_complete',
    'commercial_tobacco_alcohol_substance_use_and_impaired_driving_complete',
    'maternal_and_child_health_complete',
    'food_and_nutrition_complete',
    'adverse_child_experiences_ace_complete',
    'detailed_demographics_complete',
    'survey_experience_complete'
]
# Ensure columns exist
ordered_sections = [c for c in ordered_sections if c in df.columns]
funnel_data = (df[ordered_sections] == 2).all(axis=0).cumsum() # This is not right for funnel.
# Funnel: % of total respondents who completed up to section X
funnel_data = [(df[ordered_sections[:i+1]] == 2).all(axis=1).sum() for i in range(len(ordered_sections))]
funnel_pct = [val / len(df) * 100 for val in funnel_data]

# Date/Time Patterns
df['date_dt'] = pd.to_datetime(df['whna_2022_date'], format='mixed', errors='coerce')
df['day_of_week'] = df['date_dt'].dt.day_name()
df['hour'] = df['date_dt'].dt.hour
df['month_year'] = df['date_dt'].dt.to_period('M')

# 2.2 Mode Effects
# mode_survey: 1=Online, 2=Video? (Need to check unique values or use documentation if provided, assume 1/2)
mode_counts = df['mode_survey'].value_counts()
mode_comp_rate = df.groupby('mode_survey')['survey_experience_complete'].apply(lambda x: (x == 2).mean() * 100)

# 2.3 Item Non-Response (555 frequency)
non_resp_count = (df == 555).sum().sort_values(ascending=False)
non_resp_pct = (non_resp_count / len(df)) * 100

# --- Visualizations for Section 2 ---
plt.figure(figsize=(12, 6))
sns.barplot(x=funnel_pct, y=[s.replace('_complete', '') for s in ordered_sections], hue=[s.replace('_complete', '') for s in ordered_sections], palette='Blues_r', legend=False)
plt.title('Survey Completion Funnel (% of Total Respondents)')
plt.xlabel('Percentage Completed')
plt.savefig('section2_funnel.png')

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='day_of_week', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Survey Responses by Day of Week')
plt.savefig('section2_dow.png')
```
